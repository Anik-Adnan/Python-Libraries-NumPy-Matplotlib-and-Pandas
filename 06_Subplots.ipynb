## Visualisation in Python - Matplotlib

### Subplots: Comparing the trend of sales across different regions

-  Plotting different categories in the same graph or next to each other
-  Helps in comparing and summarising the information over multiple categories

```python
# importing numpy and the pyplot package of matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Data for sales made by the company across three regions over the years 2012 to 2015
years = np.array(['2012', '2013', '2014', '2015'])
sales_Africa = np.array([127187.27, 144480.70, 229068.79, 283036.44])
sales_Asia_Pacific = np.array([713658.22, 863983.97, 1092231.65, 1372784.40])
sales_Europe = np.array([540750.63, 717611.40, 848670.24, 1180303.95])
```

```python
# Plotting the trend of sales in African region over the 4 years
# Check the array names before plotting to avoid error
plt.plot(years,sales_Africa)
plt.show()
```

```python
# Plotting the trend of sales in all three regions over the 4 years using subplots
fig,ax = plt.subplots()

africa, = ax.plot(years,sales_Africa)
africa.set_label("Africa")

Asia, = ax.plot(years,sales_Asia_Pacific)
Asia.set_label("Africa")

Europe, = ax.plot(years,sales_Europe)
Europe.set_label("Africa")

plt.legend()
plt.show()
```

```python
# Plotting the trend of sales in all three regions over the 4 years using subplots
fig,ax = plt.subplots()

africa, = ax.plot(years,sales_Africa)
africa.set_label("Africa")

Asia, = ax.plot(years,sales_Asia_Pacific)
Asia.set_label("Africa")

Europe, = ax.plot(years,sales_Europe)
Europe.set_label("Africa")
Europe.set_dashes([2,2,5,2])
plt.legend()
plt.show()
```

#### Subplots with a shared y-axis

```python
# Plotting three charts with a shared y-axis

# Plotting the trend of sales in all three regions over the 4 years using subplots
fig,ax = plt.subplots(ncols=3)

africa, = ax[0].plot(years,sales_Africa)
africa.set_label("Africa")

Asia, = ax[1].plot(years,sales_Asia_Pacific)
Asia.set_label("Africa")

Europe, = ax[2].plot(years,sales_Europe)
Europe.set_label("Africa")
Europe.set_dashes([2,2,5,2])
# plt.legend()

plt.legend([africa, Asia, Europe],["Africa 1", "Asia 2", "Europe 3"])

plt.show()

plt.show()
```

Arrays of Plots

In plt.subplot(), the numbering starts from the top-left element and moves rightward along each row, after which it moves to the next row. Out of the following subplots, which one does NOT represent a plot that is part of the second column from the left in the array of subplots?


```python
import matplotlib.pyplot as plt
import numpy as np

# Sample Data for plotting
x = np.linspace(0, 10, 100)  # Common x values for all subplots

# Create a figure for plotting
plt.figure(figsize=(9, 9))  # Adjust figure size

# Subplot at position 2 (top-left, second column)
plt.subplot(4, 4, 2)
plt.title('subplot(4,4,2) - Second Column')
plt.plot(x, np.sin(x))

# Subplot at position 14 (bottom-left, second column)
plt.subplot(4, 4, 14)
plt.title('subplot(4,4,14) - Second Column')
plt.plot(x, np.cos(x))

# Subplot at position 7 (bottom-left, first column)
plt.subplot(3, 3, 7)  # This should be in the first column
plt.title('subplot(3,3,7) - First Column')
plt.plot(x, np.tan(x))

# Subplot at position 8 (bottom-left, second column)
plt.subplot(3, 3, 8)
plt.title('subplot(3,3,8) - Second Column')
plt.plot(x, np.exp(x))

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
