Box Plot
Find the interquartile range for the box plot created over the following list:

list_1 = [48.49, 67.54, 57.47, 68.17, 51.18, 68.31, 50.33, 66.7, 45.62, 43.59, 53.64, 70.08, 47.69, 61.27, 44.14, 51.62, 48.72, 65.11]

[Hint: use np.percentile to find the percentile values]


```python
list_1 = [48.49, 67.54, 57.47, 68.17, 51.18, 68.31, 50.33, 66.7, 45.62, 43.59, 53.64, 70.08, 47.69, 61.27, 44.14, 51.62, 48.72, 65.11]

# plotting box plot for each category
plt.boxplot(list_1)

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = np.percentile(list_1, 25)
Q3 = np.percentile(list_1, 75)

# Calculate IQR
IQR = Q3 - Q1

# adding title to the graph
plt.title("Sales across countries and  products categoris ")

# labeling the axes
plt.xlabel("Product Category")
plt.ylabel("Sales")

# Replacing the x ticks with respective category
plt.xticks((1,2,3),["Technology","Office Suppliers","Furniture"])

plt.show()

print(f"Q1 (25th Percentile): {Q1}")
print(f"Q3 (75th Percentile): {Q3}")
print(f"Interquartile Range (IQR): {IQR}")
```
