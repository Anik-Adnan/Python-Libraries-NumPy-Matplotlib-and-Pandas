# Manual Indexing in Pandas

## Description
Create a **Pandas Series** that contains the first `n` natural numbers and their respective squares.  
The first `n` numbers should appear in the **index position**.

**Hint:** Use **manual indexing**.

---

## Format

**Input:**  
A natural number `n`

**Output:**  
A pandas Series with the first `n` natural numbers as **index** and their squares as **values**.

---

## Example

**Input:**


```python
# Read the variable from STDIN
n = int(input())

import numpy as np
import pandas as pd

numbers = np.arange(1,n+1)

squares = numbers ** 2

series = pd.Series(data=squares, index=numbers)

print(series)
```

## Filtering a DataFrame Using Multiple Conditions

### Description
Create a dataframe such that:

- `Market` is 'Africa'
- `Profit` is greater than 1000
- `Sales` is greater than 200000  

All columns should be printed.

---



```python
filtered_df = sales.loc[
    (sales['Market'] == 'Africa') &
    (sales['Profit'] > 1000) &
    (sales['Sales'] > 2000)
]

print(filtered_df)
```
