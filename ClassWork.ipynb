```python
import numpy as np
import matplotlib.pyplot as plt

# Function to estimate Pi using Monte Carlo method
def estimate_pi(num_points):
    # Generate random points
    x = np.random.uniform(-1, 1, num_points)
    y = np.random.uniform(-1, 1, num_points)

    # Calculate distances from the origin
    distances = x**2 + y**2

    # Count points inside the unit circle without using np.sum
    inside_circle = 0
    for distance in distances:
        if distance <= 1:
            inside_circle += 1

    # Estimate Pi
    pi_estimate = (inside_circle / num_points) * 4
    return pi_estimate, x, y, distances

# Number of points to generate
num_points = 1000000  # Use a large number for a good estimate

# Estimate Pi
pi_estimate, x, y, distances = estimate_pi(num_points)
print(f"Estimated Pi: {pi_estimate}")

# Plot the points for visualization
num_points_plot = 1000  # Use a smaller number for plotting
x_plot, y_plot, distances_plot = x[:num_points_plot], y[:num_points_plot], distances[:num_points_plot]

# Plotting the points
fig, ax = plt.subplots()
circle = plt.Circle((0, 0), 1, edgecolor='black', fill=False)
ax.add_artist(circle)

# Color points inside the circle red, outside blue
colors = ['red' if dist <= 1 else 'blue' for dist in distances_plot]
ax.scatter(x_plot, y_plot, c=colors, s=1)
ax.set_aspect('equal', 'box')
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
plt.title(f"Monte Carlo Pi Estimation\nEstimated Pi = {pi_estimate}")
plt.show()
```

```python
import random

# Analytical Approach
def analytical_probability():
    # Probability of not rolling a double six in a single roll
    probability_single_roll_no_double_six = 35 / 36
    # Probability of not rolling a double six in 24 rolls
    probability_no_double_six_24_rolls = probability_single_roll_no_double_six ** 24
    # Probability of rolling at least one double six in 24 rolls
    probability_at_least_one_double_six = 1 - probability_no_double_six_24_rolls
    return probability_at_least_one_double_six

# Simulation Approach
def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)

def simulate_double_six(num_simulations, num_rolls):
    double_six_count = 0

    for _ in range(num_simulations):
        found_double_six = False
        for _ in range(num_rolls):
            die1, die2 = roll_dice()
            if die1 == 6 and die2 == 6:
                found_double_six = True
                break
        if found_double_six:
            double_six_count += 1

    probability = double_six_count / num_simulations
    return probability

# Main function to run both approaches
def main():
    # Analytical probability
    analytical_prob = analytical_probability()
    print(f"Analytical probability of getting at least one double six in 24 rolls: {analytical_prob:.4f}")

    # Simulation parameters
    num_simulations = 100000
    num_rolls = 24

    # Simulation probability
    simulation_prob = simulate_double_six(num_simulations, num_rolls)
    print(f"Estimated probability from simulation of getting at least one double six in 24 rolls: {simulation_prob:.4f}")

# Run the main function
if __name__ == "__main__":
    main()
```

```python
import random

def simulate_double_six(num_simulations, rolls_per_simulation):
    success_count = 0

    for _ in range(num_simulations):
        for _ in range(rolls_per_simulation):
            die1 = random.randint(1, 6)
            die2 = random.randint(1, 6)
            if die1 == 6 and die2 == 6:
                success_count += 1
                break

    return success_count / num_simulations

# Parameters
num_simulations = 100000  # Number of simulations
rolls_per_simulation = 24  # Number of rolls per simulation

probability = simulate_double_six(num_simulations, rolls_per_simulation)
print("Estimated probability of getting at least one double six in 24 rolls:", probability)
```

```python
import scipy.stats as stats

# Observed frequencies
observed = [[18, 36, 21, 9, 6],
            [12, 36, 45, 36, 21],
            [6, 9, 9, 3, 3],
            [3, 9, 9, 6, 3]]

# Calculate expected frequencies
expected = []
for i in range(len(observed)):
    row = []
    for j in range(len(observed[0])):
        row.append((sum(observed[i]) * sum(row[j] for row in observed)) / sum(sum(row) for row in observed))
    expected.append(row)

# Calculate chi-square statistic
chi2, p_value, dof, expected = stats.chi2_contingency(observed)

# Print results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("Critical Value at =0.05 with df=8:", 21.03)
print("P-value:", p_value)
print("Expected Frequencies:")
for i in range(len(expected)):
    print(expected[i])
```

```python
import numpy as np
import scipy.stats as stats
import math

# Given data
data = [67, 63, 33, 69, 53, 51, 49, 78, 77, 83, 47, 53, 51, 49, 78, 77,
        83, 47, 67, 63, 33, 69, 53, 51, 49, 78, 77, 83, 47, 53, 51, 77]

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Determine the number of bins using Sturges' rule
num_bins = math.ceil(math.log2(len(data)) + 1)

# Calculate the bin edges
bin_edges = np.linspace(min(data), max(data), num_bins + 1)

# Calculate observed frequencies
observed_freq, _ = np.histogram(data, bins=bin_edges)

# Calculate expected frequencies assuming normal distribution
expected_freq = []
for i in range(num_bins):
    lower_bound = bin_edges[i]
    upper_bound = bin_edges[i + 1]
    prob = stats.norm.cdf(upper_bound, mean, std_dev) - stats.norm.cdf(lower_bound, mean, std_dev)
    expected_freq.append(prob * len(data))

# Calculate Chi-square statistic
chi_square_stat = sum((o - e) ** 2 / e for o, e in zip(observed_freq, expected_freq))

# Degrees of freedom: number of bins - 1 - number of estimated parameters (mean, std_dev)
dof = num_bins - 1 - 2

# Critical value for chi-square distribution at 24% significance level
critical_value = stats.chi2.ppf(0.76, dof)  # 1 - 0.24 = 0.76

# Print results
print("Chi-square Statistic:", chi_square_stat)
print("Degrees of Freedom:", dof)
print("Critical Value at 24% significance level:", critical_value)

# Conclusion
if chi_square_stat < critical_value:
    print("Fail to reject the null hypothesis: Data fits the normal distribution")
else:
    print("Reject the null hypothesis: Data does not fit the normal distribution")
```

```python
import scipy.stats as stats

def poisson_probability(avg, est):
    # Calculate the Poisson probability using scipy
    poisson_prob = stats.poisson.pmf(est, avg)
    return poisson_prob

# Input values
cases = [(250, 275), (300, 350), (100, 150)]

# Calculate and print the Poisson probabilities
for i, (avg, est) in enumerate(cases):
    prob = poisson_probability(avg, est)
    print(f"Case {i + 1}: {avg}, {est} {prob:.6f}")
```

```python
import math

def poisson_probability(lamda, k):
    # Calculate ln(lamda^k)
    ln_lamda_power_k = k * math.log(lamda)

    # Calculate ln(k!)
    ln_k_factorial = math.log(math.factorial(k))

    # Calculate ln(P(X = k))
    ln_poisson_prob = ln_lamda_power_k - lamda - ln_k_factorial

    # Calculate P(X = k)
    poisson_prob = math.exp(ln_poisson_prob)

    return poisson_prob

# Input values
cases = [(250, 275), (300, 350), (100, 150)]

# Calculate and print the Poisson probabilities
for i, (avg, est) in enumerate(cases):
    prob = poisson_probability(avg, est)
    print(f"Case {i + 1}: {avg}, {est} {prob:.6f}")
```
